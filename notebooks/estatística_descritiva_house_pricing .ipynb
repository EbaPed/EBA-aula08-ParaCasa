### **Relembrando algumas medidas estatísticas importantes:**

As medidas de tendência central são:

- **Media**

Dados os $n$ números ${\{x_{1}, x_{2}, \ldots, x_{n}\}}$, a média aritmética é definida como:

$$
\begin{equation}
    \bar {x} = {\frac{1}{n}} \sum_{i = 1}^{n}x_{i} ={\frac{x_{1} + x_{2} + \cdots + x_{n}}{n}}
\end{equation}
$$

- **Mediana**

Sejam ${x_{1}, x_{2}, x_{3}, \ldots, x_{n}}$ os dados de uma amostra ordenada em ordem crescente e designando a mediana como $M_{e}$, distinguimos dois casos:

a) Se $n$ for ímpar, a mediana é o valor que ocupa a posição $\frac{(n + 1)}{2}$ uma vez ordenados os dados, em ordem crescente ou decrescente, pois este é o valor central. Ou seja: $M_{e}=x_{{(n+1)/2}}$.

b) Se $n$ for par, a mediana é a média aritmética dos dois valores centrais. Quando $n$ é par, os dois dados no centro da amostra ocupam as posições $\frac{n}{2}$ e $ \frac{n}{2} + 1$. Ou seja: $M_{e}=(x_{{{n / 2}}} + x_{{{{(n / 2)}}+1}})/2$.

-  **Moda**

A moda é o valor dos dados com mair freqüência em um conjunto.



*É importante ressaltar que com o bom uso da estatística evita quaisquer tipo de generalização ou inferência além do que está disponível. Isso significa que as estatísticas descritivas são apenas a representação dos dados (amostra) disponíveis e não se baseiam em nenhuma teoria de probabilidade.*

**Amplitude interquartílica:**	A distância entre o 1o. e o 3o. quartis (Q3-Q1); assim, ele ultrapassa o meio de 50% dos dados.

## **Dispersão**

Estamos agora interessandos em entender o comportamento perante a distribuição de features e/ou target numéricos. É importante responder, por exemplo, se:

1) A distribuição é normal? Essa pergunta deve ser feita pois diversas outras estatísticas que vocês verão nos próximos módulos são aplicadas *apenas* em distribuições normais (DISTRIBUIÇÕES PARAMETRICAS)

2) Qual é o desvio padrão da distribuição? Será que os dados variam muito?

3) Qual seria a curtose dessa distribuição? Com isso, conseguiremos entender se existem mais ou menos dados concentrados na cauda

4) Se não for normal, qual é a skew dessa distribuição? Com isso, podemos entender se a cauda é desviada mais pra direita ou esqueda



Vamos analisar nosso KPI

````python
#Sale
plt.hist(df['SalePrice'], bins =50)
````

Claramente a distribuição apresenta uma cauda longa, não sendo normal portanto. Vemos a cauda deslocada para a direita, mostrando que temos poucos valores com um preço muito mais alto (outliers!) - lembrem que vimos isso também pelo nosso boxplot na aula passada

Agora vamos ver o desvio padrão da amostra

````python
desvio = df['SalePrice'].std()
desvio
````

Lembram que nossa média era 180.9k? Pois então, estamos vendo um desvio padrão de 79.4k, que parece ser bastante alto a julgar pelo valor da média. Vamos ver isso percentualmente pelo coeficiente de variancia:

````python
coef_var =  (desvio/df['SalePrice'].mean())*100
coef_var
````

O coeficiente de variancia é de 43.9%, ou seja, nosso desvio padrão representa 43.9% da nossa média.

Uma interpretação intuitiva para o desvio padrão é entendê-lo com um uma medida de erro caso aproximarmos (ou estimarmos) todos os valores da série dados pela sua média. Se nos basearmos somente pela média desses dados, entendendo que o desvio padrão é bastante alto, corremos o risco de interpretar mal esse dado. Daí a importancia das medidas que vimos até agora: mediana, moda e o próprio desvio padrão.

## 1) A qualidade da casa é superior para as casas caras?

Para usar o teste t comparando a média de OverallQual para as casas caras e baratas precisamos antes de mais nada descobrir se as variancias são iguais, pois temos testes t diferentes para cada caso!

````python
df
````

````python
df["categoria_preco"].value_counts()
````

Temos mais de 20 observações em cada, portanto podemos usar o teste F

H0: Variancias iguais
Ha: Variancias diferentes

````python
df.groupby("categoria_preco")["OverallQual"].var()
````

````python
import scipy.stats

#qualidade geral apenas das caras
caras = df[df.categoria_preco == "cara"].OverallQual

#qualidade geral apenas das baratas
baratas = df[df.categoria_preco == "barata"].OverallQual

df_cara = len(caras) - 1 #grau de liberdade da primeira amostra

df_barata = len(baratas) - 1 #grau de liberdade da segunda amostra

F = baratas.var() / caras.var() #A maior variância tem que ir no numerador!

alpha = 0.05

#p-valor para o teste F
1 - scipy.stats.f.cdf(F, df_barata, df_cara)
````

p-valor é maior do que alpha (0.05), portanto podemos considerar as variancias iguais!

### Agora podemos partir para a análise de qualidade por preço

Verificando se podemos aplicar o teste t

Temos mais de 30 amostras e queremos comparar médias, portanto podemos aplicar o teste t 2 samples para amostras independentes com desvio-padrao populacional desconhecido e variancias iguais

H0: médias iguais

Ha: médias diferentes

````python
import scipy.stats as s

s.ttest_ind(a=caras, b=baratas, equal_var=True) #vimos anteriormente que as variancias sao iguais

#retorna t e p-valor
````

O p-valor é muito pequeno, podemos afirmar a 95% de confiança (e até muito mais) que as casas tem de fato uma média diferente.

A ANOVA preve que podemos verificar se as médias se diferem se:


* Amostras aleatórias
* Grupos independentes
* A variável dependente é contínua
* Amostras tem pelo menos n = 15 cada

### Bartlett

* A hipótese nula para o teste é que as variâncias são iguais para todas as amostras. Em termos estatísticos, isso é:
H0: σ12 = σ22 = … = σk2.

* A hipótese alternativa é que as variâncias não são iguais para um par ou mais:
Ha: σ12 ≠ σ22 ≠… ≠ σk2.
* Pode ser usada se tiver dados normalmente distribuídos ou +20 amostras


````python
import scipy.stats as stats

stats.bartlett(list_RL, list_C, list_FB, list_RH, list_RM)
````

Como p-valor é menor que alpha, concluimos que elas não tem a mesma variancia. Portanto, poderiamos usar a ANOVA de Welch

### ANOVA WELCH
* A ANOVA de Welch compara duas médias para ver se elas são iguais. É uma alternativa à ANOVA clássica e pode ser usada mesmo que seus dados violem a suposição de homogeneidade de variâncias.
* Pode ser usada se tiver dados normalmente distribuídos ou +15 amostras

````python
! pip install pingouin
````

````python
import pingouin as pg

pg.welch_anova(dv='SalePrice', between='MSZoning', data=df)
````

pvalor é menor do que alpha (0.05), portanto, rejeitamos H0 e concluimos que a 95% de confiança, todas as médias não são iguais

### Porém, aqui cometemos um abuso!

ANOVA Welch e o Bartlett requer que os dados sejam normais ou que tenhamos pelo menos 15 dados em cada grupo. O grupo C tem apenas 10.

Para esse teste, precisaríamos de um teste **não paramétrico** para de fato comprovar essa diferença, como o Kruskal–Wallis

**Kruskal–Wallis**
O teste Kruskal-Wallis é um teste não paramétrico que pode ser usado para determinar se existem diferenças estatisticamente significativas entre dois ou mais grupos de uma variável independente em um variável dependente contínua ou ordinal. É considerada a alternativa não paramétrica à ANOVA one-way e uma extensão do teste U de Mann-Whitney para permitir a comparação de mais de dois grupos independentes.

````python
stats.kruskal(list_RL, list_C, list_FB, list_RH, list_RM)
````

pvalor é menor do que alpha (0.05), portanto, rejeitamos H0 e concluimos que a 95% de confiança, todas as médias não são iguais

**Post Hoc**

Se p-valor for menor do que 5%, rejeitamos H0 e, portanto, dizemos que os pares são diferentes

````python
#teste nao parametrico games howell
pg.pairwise_gameshowell(dv='SalePrice', between='MSZoning', data=df)
````

### O que é Multicolinearidade?

Multicolinearidade é um fenômeno que ocorre quando duas ou mais variáveis independentes em um modelo de regressão estão altamente correlacionadas. Isso significa que uma variável pode ser predita linearmente a partir das outras com um alto grau de precisão. Em termos simples, é quando as variáveis explicativas fornecem informações redundantes umas sobre as outras.

### Por que a Multicolinearidade é um Problema?

A multicolinearidade pode causar diversos problemas na análise de regressão, tais como:

1. **Coeficientes Instáveis**: Pequenas alterações nos dados podem resultar em grandes variações nos coeficientes das variáveis independentes.
2. **Dificuldade em Determinar o Efeito Individual das Variáveis**: Quando as variáveis são altamente correlacionadas, torna-se difícil isolar o efeito de uma variável individual no resultado.
3. **Altos Erros Padrão**: Isso pode levar a intervalos de confiança mais amplos e a testes estatísticos menos precisos.

### Como Checar a Multicolinearidade

Existem várias maneiras de detectar a multicolinearidade, incluindo:

1. **Matriz de Correlação**: Examinar a matriz de correlação entre as variáveis independentes. Correlações altas (próximas de 1 ou -1) podem indicar multicolinearidade.

2. **Variance Inflation Factor (VIF)**: O VIF quantifica o grau de multicolinearidade. Um VIF superior a 10 é geralmente considerado indicativo de alta multicolinearidade, mas em algumas áreas, um valor acima de 5 já pode ser considerado preocupante.

3. **Eigenvalues and Condition Number**: Análise dos autovalores da matriz de correlação das variáveis independentes. Um número de condição elevado (acima de 30) pode sugerir problemas de multicolinearidade.


### Soluções para Multicolinearidade

- **Remover Variáveis**: Remover uma ou mais variáveis altamente correlacionadas.
- **Combinar Variáveis**: Criar uma nova variável a partir da combinação das variáveis correlacionadas.
- **Análise de Componentes Principais (PCA)**: Transformar as variáveis correlacionadas em um novo conjunto de variáveis não correlacionadas.

### O que é Centralização de Variáveis Numéricas?

Centralizar variáveis numéricas é uma técnica usada em análise de dados e modelagem estatística que envolve subtrair a média de cada variável de seus valores individuais. Isso resulta em uma nova variável cuja média é zero. A centralização é uma forma de pré-processamento que pode melhorar a interpretação e a performance de certos modelos estatísticos.

### Por Que Centralizar Variáveis?

1. **Interpretação**: Em modelos de regressão, centralizar as variáveis pode facilitar a interpretação dos coeficientes. Por exemplo, quando as variáveis são centralizadas, o intercepto representa o valor médio da variável dependente quando todas as variáveis independentes estão em seus valores médios.

2. **Convergência de Algoritmos**: Alguns algoritmos de otimização convergem mais rapidamente e de forma mais estável quando as variáveis são centralizadas.

3. **Multicolinearidade**: Centralizar pode ajudar a reduzir a multicolinearidade, especialmente em interações entre variáveis.

4. **Escalas de Variáveis**: Em alguns casos, centralizar pode ajudar a lidar com variáveis em escalas muito diferentes, embora a padronização (centralização seguida de divisão pelo desvio padrão) seja uma técnica mais comum para isso.

Lembrando que aqui não é necessário centralizar as dummies

````python
import statsmodels.formula.api as smf
import statsmodels.api as sm

mod1 = smf.ols(formula = 'SalePrice ~ OverallQual + OverallCond + BedroomAbvGr + TotRmsAbvGrd + GrLivArea + TotalBsmtSF + YearBuilt + MSZoning_FV + MSZoning_RH + MSZoning_RL + MSZoning_RM + BldgType_2fmCon + BldgType_Duplex + BldgType_Twnhs + BldgType_TwnhsE', data = df_center)
res = mod1.fit()
````
